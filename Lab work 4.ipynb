3. Вивести середнє значення та відхилення курсу за кожний місяць.

```python
import datetime
import math
import requests
import locale

locale.setlocale(locale.LC_TIME, 'uk_UA.utf8')

#main function for gathering rates from the given periods and calculation on them
def gathering_info(year_: int, month_: int, day_: int, range_of_days: int) -> dict:
    """Take start date and the range of days of the necessary period as a tuple 
    and return mean currency rate and standard deviation from this period as floats
    in form of a dictionary with a month and year as a key and a value
    consisting of another dictionary with mean rate and standard deviation as keys.
    : params `year_, month_, day_`: start date of the necessary period input as 
    integers in format 'year, month, day'
    : param `range_of_days`: amount of days of the necessary period input as an integer
    """
    list_month_rate = []
    start = datetime.date(year_, month_, day_)
    for i in range(range_of_days):
        date = start + datetime.timedelta(days=i)
        params = {
            "date": date.strftime("%Y%m%d"),
            "json": "json"
        }
        r = requests.get(
            'https://bank.gov.ua/NBUStatService/v1/statdirectory/exchange',
            params=params).json()
        for x in r:
            if x['cc'] == 'USD':
                x['rate'] = float(x['rate'])
                list_month_rate.append(x['rate'])

        mean = round(sum(list_month_rate)/len(list_month_rate), 4)
    
        n = len(list_month_rate)
        mean = sum(list_month_rate) / n
        variance = sum([((num - mean) ** 2) for num in list_month_rate]) / n
        std_dev = round(math.sqrt(variance), 4)
    
    dict_USD_rate = {}
    month_name = start.strftime('%B')
    month_name_decoded = month_name.encode('cp1251').decode('utf-8')
    year = date.strftime('%Y')
    data_period = month_name_decoded + " " + year
    dict_USD_rate[data_period] = {
    'середній курс USD': mean,
    'відхилення курсу USD': std_dev
    }
    
    return dict_USD_rate
    

# Periods which we want to research
list_of_dates = [(2022, 3, 1, 31), (2022, 4, 1, 30), (2022, 5, 1, 31), (2022, 6, 1, 30),
                       (2022, 7, 1, 31), (2022, 8, 1, 31), (2022, 9, 1, 30), (2022, 10, 1, 31),
                       (2022, 11, 1, 30), (2022, 12, 1, 31), (2023, 1, 1, 31), (2023, 2, 1, 28)]

# create a list for the entire researched period to make easier pickling further
result = []
for t in list_of_dates:
    result.append(gathering_info(*t))

print(result)


```

<b> Задание: 29. В предположении, что приход и расход денег на каждом пункте приема фиксируется не чаще одного раза в день

Использовать таблицы Income_o и Outcome_o.

(первичным ключом в таблицах является столбец code), требуется получить таблицу, в которой каждому пункту за каждую дату 
выполнения операций будет соответствовать одна строка. Вывод: point, date, суммарный расход пункта за день (out), 
суммарный приход пункта за день (inc). Отсутствующие значения считать неопределенными (NULL).


```python
SELECT point, date, SUM(sum_out), SUM(sum_inc) 
FROM
(SELECT point, date, SUM(inc) AS sum_inc, NULL AS sum_out 
FROM Income 
GROUP BY point, date 
UNION
SELECT point, date, NULL AS sum_inc, SUM(out) AS sum_out 
FROM
Outcome 
GROUP BY point, date) AS a 
GROUP BY point, 
date ORDER BY point
```
