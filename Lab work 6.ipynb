```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
xls = pd.ExcelFile('FOR_LAB_6.xlsx')
dfO = pd.read_excel(xls, 'Orders')
dfC = pd.read_excel(xls, 'Customers')
dfP = pd.read_excel(xls, 'Product')
```

```python
# dfO.head()
```

```python
# dfC.head()
```

```python
df_new = dfO.merge(dfC, how='inner', on ='Customer ID')
df_new.head(3)
```

```python
# Because the plot in Matplotlib and Seaborn cannot be so wonderfully interactive as in PowerBI, 
# I choosed the top 10 customers from the whole period for the main plot

# Sales by Customers Plot

top_customers = df_new['Customer Name'].value_counts().sort_values(ascending=False).head(10).index
top_customers
```

```python
sales_by_customer = (df_new.groupby(['Customer Name'])['Sales'].aggregate('sum')).sort_values(ascending=False).head(10).index
sales_by_customer
```

```python
sales_by_top_customer = (df_new[df_new['Customer Name'].isin(sales_by_customer)]
                                   .groupby(['Customer Name'])[['Sales']]
                                   .aggregate('sum')).sort_values(by = 'Sales')
```

```python

axes = sales_by_top_customer.plot(kind='barh', figsize=(8,8), legend = False, color='lightseagreen')
axes.set_title("Sales by customer", fontsize = 18)
axes.set(ylabel=None)
sns.set_style("whitegrid", {"grid.linestyle": "dotted"});
```

```python
# Time of Delivery Plot

import datetime
from datetime import datetime, timedelta


df_new['Time Delivery'] = (df_new['Ship Date'] - df_new['Order Date']).dt.days
df_new.head(3)
```

```python
av_delivery = (df_new.groupby('Ship Mode')['Time Delivery'].mean()
)
# av_delivery
ax = av_delivery.plot(kind='bar', figsize=(8,4), color = 'lightseagreen')
sns.set_style("whitegrid", {"grid.linestyle": "dotted"})
ax.set_title("Average Time of Delivery", fontsize = 16)
ax.set(xlabel=None)
plt.xticks(rotation=360);
```

```python
# Sales by Month Plot (June)

df2 = df_new[df_new['Order Date'].dt.strftime('%Y-%m') == '2015-06']
df2.head(3)
```

```python

sales_june = df2.groupby(['Order Date']).agg(overall_sales= ('Sales' , 'sum'))
# sales_june
```

```python
ax = sales_june.plot(figsize=(8, 4), legend = None, color = 'lightseagreen')
ax.set_title("Time Series", fontsize = 14)
ax.set(xlabel=None);
```

```python
# Shipping Price for Donut Chart

ship_price = df_new.groupby(['Ship Mode']).agg(shipping_price = ('Shipping Cost', 'sum'))
# ship_price
```

```python

ship_price_percents = ship_price.div(ship_price['shipping_price'].sum(), axis=0)
ship_price_percents
```

```python
costs = ship_price_percents.shipping_price.values.tolist()
# costs
```

```python
colors = ['lightseagreen', 'skyblue', 'olive']
labels = ['Delivery Truck', 'Express Air', 'Regular Air']
plt.pie(costs, 
        autopct='%1.1f%%', colors = colors, labels = labels)

 
# draw circle
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
 
# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)
 
# Adding Title of chart
plt.title('Shipping Cost by Shipping Mode');
```
