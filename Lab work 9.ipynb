```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
from os import listdir
from numpy import asarray
from numpy import save
from random import sample
import keras
import keras.preprocessing.image
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
#Because of the different size of all pictures, we should firstly standartize them
#A good decision - we can do this by transforming all pics to one shape/size and to create array from them
```

```python
# Define path to the dataset folder
folder = r'C:\\Users\\liman\\Pictures\\train\\'

#I did not take all the pictures into work because my RAM did not allow that
# Specify the number of images for training and testing
num_train_images = 20000
num_test_images = 2000

# Initialize empty lists for storing photos and labels
train_photos = []
train_labels = []
test_photos = []
test_labels = []

# Enumerate files in the directory
file_list = os.listdir(folder)

# Randomly select a subset of files for training
train_files = sample(file_list, num_train_images)

# Process the training files
for file in train_files:
    # Determine class
    output = 0.0
    if file.startswith('dog'):
        output = 1.0
        
    # Load image
    photo = load_img(os.path.join(folder, file), target_size=(100, 100))
    
    # Convert to numpy array
    photo = img_to_array(photo)
    
    # Store photo and label in the training sets
    train_photos.append(photo)
    train_labels.append(output)

# Randomly select a subset of files for testing
# Exclude the files used for training
test_files = sample(set(file_list) - set(train_files), num_test_images)

# Process the testing files
for file in test_files:
    # Determine class
    output = 0.0
    if file.startswith('dog'):
        output = 1.0
        
    # Load image
    photo = load_img(os.path.join(folder, file), target_size=(100, 100))
    
    # Convert to numpy array
    photo = img_to_array(photo)
    
    # Store photo and label in the testing sets
    test_photos.append(photo)
    test_labels.append(output)

# Convert photos and labels to numpy arrays
train_photos = np.array(train_photos)
train_labels = np.array(train_labels)
test_photos = np.array(test_photos)
test_labels = np.array(test_labels)

print("Train set shape:", train_photos.shape, train_labels.shape)
print("Test set shape:", test_photos.shape, test_labels.shape)
```

```python
# save the reshaped photos
save('train_dogs_vs_cats_photos.npy', train_photos)
save('train_dogs_vs_cats_labels.npy', train_labels)

save('test_dogs_vs_cats_photos.npy', test_photos)
save('test_dogs_vs_cats_labels.npy', test_labels)
```

```python
# Define the model architecture
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(100, 100, 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(train_photos, train_labels, epochs=10, batch_size=32)

# Evaluate the model on the test set
loss, accuracy = model.evaluate(test_photos, test_labels)
print("Test Loss:", loss)
print("Test Accuracy:", accuracy)
```
